### question_1

برای اعمال یک عملیات بر روی کل سطر ها میتوان از سینتکس زیر استفاده کرد که برای هر سطر به صورت جداگانه اعمال میشود

### question_2

* `df['mean']` =  اضافه کردن سطر جدید

* `df.loc[:,'2022':'1970']` =  جدا کردن ستون های مورد نظر

* `mean(axis=1)` = میانگین گرفتن سطری مقادیر

```python
df['mean'] = df.loc[:,'2022':'1970'].mean(axis=1)
df
```

### question_3

* `for i in df.loc[:,'2022':'1970']:` = انتخاب سطر های مورد نظر
* `df[i+'%']` = اضافه کردن سطر جدید با اسم سال و درصد
* `df[i]*100 /df[i].sum()` = محاسبه درصد جمعیت یک سال یک کشوراز جمعیت کل جهان در همان سال

```python
for i in df.loc[:,'2022':'1970']:
    df[i+'%'] = df[i]*100 /df[i].sum()
df
```

### question_4

```python
df['growth rate'] = df['2022']/df['2020']
df
```

### question_5_6

* `sort_values(by='2020%')` = این تابع دیتافریم را سورت میکند و پارامتر ورودی آن نام ستونی که طبق آن سورت میشود را میگیرد
* `ascending=False` = این پارامتر صعودی بودن سورت را مشخص میکند
* `.index` = این پارامتر اندیس دیتافریم مرتب شده را برای هر سطر برمیگرداند که آن را در ستون جدیدی ذخیره کردم. از طرفی خود تابع سورت تغییرات را مستقیم روی دیتافریم اصلی ذخیره نمیکند پس بعد از اتمام این تابع دوباره اندیس های اصلی بر اساس حالت اولیه هستند و نتیجه سورت شده نیست.صرفا اندیس حالت سورت شده ذخیره شده است. 

```python
df['Rank'] = df.sort_values(by='2020%', ascending=False).index
df
```

### question_7

* `df['2022'] > 300_000` = این عبارت به ازای هر سطر مقدار `true` یا `false` برمیگرداند
* `.replace(False, 'less than 300'` = این تابع دو پارامتر ورودی میگیرد که هرکجا در ستون مشخص شده مقدار معادل پارامتر اول ورودی را ببیند مقدار پارامتر دوم را با ان جایگزین میکند
* `inplace=True` = این پارامتر جایگذاری را در حالت درجا انجام میدهد که یعنی خود دیتافریم اصلی تغییر میکند
* همین کار را برای بیشتر از ۳۰۰ میلیون هم انجام میدهیم.فقط در دیتاست ورودی جمعیت کشور ها با ۳ تا صفر کمتر وجود دارد که به جای ۳۰۰،۰۰۰،۰۰۰ باید با عبارت ۳۰۰،۰۰۰ مقایسه کنیم

```python
df['binary population class'] = df['2022'] > 300_000
df['binary population class'].replace(False, 'less than 300', inplace=True)
df['binary population class'].replace(True, 'more than 300', inplace=True)
df
```

### question_8

* `df.loc[df['2022%']>5,` = جدا کردن سطر هایی که درصد جمعیت بیشتر از ۵ دارند
* `'population class'] = 2` = انتخاب ستون مورد نظر و مقدار دهی به آن سلول خاص
* همین کار را برای ۳ مقدار انجام میدهیم.در سوال ۲ بخش را گفته بود با مقدار ۱ پر کنیم که من فکر کنم اشتباه شده و مقدار سوم را ۲ گذاشتم

```python
df.loc[df['2022%']>5,'population class'] = 2
df.loc[df['2022%']<5,'population class'] = 1
df.loc[df['2022%']<2,'population class'] = 0
df.sort_values(by='2022%', ascending=False)
```

### question_9

* `df.loc[df['2022'] > df['1970'],'growth status'] = 0` = این خط کد مقدار ستون جدید را برای سطر هایی که جمعیت ۲۰۲۲ شان از ۱۹۷۰ شان بیشتر است برابر ۰ میگذارد. و چون ستون جدید اضافه شده باقی سطر ها که مقدار نمیگیرند میشنود np.nan
*  `df['growth status'].replace(np.nan, 1, inplace=True)` = حال برای اینکه باقی سطر ها که ۰ نبودند را برابر ۱ قرار دهیم کافیست مقادیر خالی جدول را با عدد ۱ جایگزین کنیم
* `df.sample(10)` = این عبارت هم برای نمایش ۱۰ نمونه رندوم از دیتافریم است که صحت عملکردمان را با ان چک کنیم

```python
df.loc[df['2022'] > df['1970'],'growth status'] = 0
df['growth status'].replace(np.nan, 1, inplace=True)
df.sample(10)
```

### question_10

* `cn_df = df.loc[df['CCA3'] == 'CN','2022':'1970']` = انتخاب دیتای کشور چین
* `cn_df = cn_df.transpose()` = برای کشیدن نمودار نیاز است دیتافریم به صورت ترانهاده دیتایی که داریم ذخیره شده باشد پس ترانهاده اش را حساب میکنیم

```python
cn_df = df.loc[df['CCA3'] == 'CN','2022':'1970']
cn_df = cn_df.transpose()
cn_df
```

* `fig = plt.figure(figsize=(10, 10))` = برای استفاده از ۲ نوع نمودار یک فضا مشخص کرده و ۲ زیرنمودار با مختصات و اندازه مشخص به آن اضافه میکنیم
* `plt.sca(ax1)` = این عبارت برای مقدار دهی به یک زیرنمودار خاص است
* در ادامه با مشخص کردن رنگ و نوع نمودار و عنوان دادن به آن نمودار رسم میشود

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_axes([0, 0, 0.45, 0.45])
ax2 = fig.add_axes([0.55, 0, 0.45, 0.45])

plt.sca(ax1)
scatter = plt.scatter(cn_df.index, cn_df.values, marker='*', c='r')
plt.title('scatter plot')
plt.xlabel('year')
plt.ylabel('population')

plt.sca(ax2)
plot = plt.plot(cn_df, c='b')
plt.title('plot')
plt.xlabel('year')
plt.ylabel('population')

plt.show()
```
