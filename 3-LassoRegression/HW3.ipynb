```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# EDA

```python
sales = pd.read_csv("sales_train.csv")
sales = sales.drop('date',axis=1).drop('item_price',axis=1).rename(columns={'date_block_num':'month','item_cnt_day':'cnt'})
sales
```

```python
test_df = pd.read_csv('test.csv')
test_df
```

```python
df = sales.groupby(['shop_id','item_id','month']).sum()
df
```

```python
# next  :  fillna with mean of other shops per item -> feature engineering
df = pd.pivot_table(df,index=['shop_id','item_id'], values='cnt',columns=['month']).fillna(0)
df
```

```python
test = pd.merge(test_df,df,how='left',on=['shop_id','item_id']).fillna(0)
test
```

```python
x_test = test.drop(['ID','shop_id','item_id',0],axis=1)
x_test
```

```python
df = df.loc[:,0:33].reset_index(drop=True)
df
```

```python
x_train = df.iloc[:340_000,:-1]
y_train = df.iloc[:340_000,-1:]
x_validation = df.iloc[340_000:,:-1]
y_validation = df.iloc[340_000:,-1:]
print('x_train:     ',x_train.shape)
print('y_train:     ',y_train.shape)
print('x_validation:',x_validation.shape)
print('y_validation:',y_validation.shape)
print('x_test:      ',x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
linear = LinearRegression(normalize=True)
linear.fit(x_train,y_train)
y_pred = linear.predict(x_validation)
print('score: ',linear.score(x_validation,y_validation))
print('RMSE: ',mean_squared_error(y_pred, y_validation, squared=False))
```

رگرسیون لاسو نمونه ای از رگرسیون منظم است.منظم سازی یکی از روش های مقابله با اورفیت شدن مدل با اضافه کردن اطلاعات اضافی و در نتیجه کوچک کردن مقادیر پارامتر های مدلی برای کم کردن خطا است.

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
lasso = Lasso(alpha=0.01, normalize=True)
lasso.fit(x_train,y_train)
y_pred2 = lasso.predict(x_validation)
print('score: ',lasso.score(x_validation,y_validation))
print('RMSE: ',mean_squared_error(y_pred2, y_validation, squared=False))
```

برای ارزیابی مدل های رگرسیون و مخصوصا رگرسیون خطی از معیار های زیر استفاده میشود.
